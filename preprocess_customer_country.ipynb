```python
import pandas as pd
```

```python
TRAIN_FILE = "./data/train.csv"
TEST_FILE = "./data/test.csv"

df_train = pd.read_csv(TRAIN_FILE)
df_test = pd.read_csv(TEST_FILE)
```

```python
# country 컬럼 전처리: 국가 코드로 변경
from collections import defaultdict
import pycountry

def preprocess_country():
    preprocess = defaultdict(int)
    preprocess["NaN"] = preprocess["NOT_FOUND"] = 0    # init
    missing_value = list()

    for country in df_train['customer_country']:
        if isinstance(country, float):
            preprocess["NaN"] += 1
            continue

        # last word based on '/'
        country = country.split("/")[-1]
        
        # return is None or Country object
        country = pycountry.countries.get(name=country)

        # alpha_2: 2자리 국가 코드
        # alpha_3: 3자리 국가 코드
        # name: 국가 이름
        # numeric: 국가 번호
        # official_name: 정식 국가 이름
        if country:
            code = country.alpha_3
            preprocess[code] += 1
        else:
            missing_value.append(country)
            preprocess["NOT_FOUND"] += 1

    return missing_value, preprocess
```
